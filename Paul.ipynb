# Metric: Amount of Fast Food Restaurants

As enjoyable and cheap fast food is, it isn't exactly great for your health. Therefore one of our metrics is the town with the most fast food places in their zip code. The dataset I chose was "Allegheny County Fast Food Establishments" and it contained a zipcode attached to each establishment. All I then needed to do was see what frequency of Zip code was the highest.

```python
import pandas as pd
import matplotlib.pyplot as plt

ffData = pd.read_csv("fastfood.csv")
ffData = ffData[ffData['Category'] != 'Breakfast, Drink, Other']
zipcodes = ffData['ZIP Code']
zcCounts = zipcodes.value_counts()

split_index = len(zcCounts) // 2
zipcode_counts_part1 = zcCounts.iloc[:split_index]
zipcode_counts_part2 = zcCounts.iloc[split_index:]

# First group of zip codes
plt.figure(figsize=(10, 6))
zipcode_counts_part1.plot(kind='bar')
plt.title('Frequency of Zip Codes in Fast Food Locations (Part 1)')
plt.xlabel('Zip Code')
plt.ylabel('Amount of Restaurants')
plt.grid(axis='y')  
plt.tight_layout()  
plt.show()

# Second group of zip codes
plt.figure(figsize=(10, 6))  #
zipcode_counts_part2.plot(kind='bar')
plt.title('Frequency of Zip Codes in Fast Food Locations (Part 2)')
plt.xlabel('Zip Code')
plt.yticks([0,5,10,15,20,25,30,35])
plt.ylabel('Amount of Restaurants')
plt.grid(axis='y')  
plt.tight_layout()  
plt.show()

```

From the graph we can see that area codes 15146, 15207, and 15205 are the neighborhoods with the most amount of fast food restaurants in the area at 34, 31 and 28.

## That leaves us with the neighborhoods: Monroeville, Hazelwood, and Crafton according to the US Postal Service zipcode finder as having the most amount of fast food restaurants.
