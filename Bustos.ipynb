### Data and Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

# Getting trees per neighborhood
tree_data = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da", low_memory=False)

# Counting occurrences of each neighborhood
neighborhood_counts = tree_data['neighborhood'].value_counts()

# Sorting neighborhoods by their counts in ascending order
sorted_counts = neighborhood_counts.sort_values()

# Getting the neighborhoods with the least amount of trees
least_trees_10 = sorted_counts.head(10)

# Grouping data by neighborhood and calculating average tree height and width
neighborhood_avg_size = tree_data.groupby('neighborhood')[['height', 'width']].mean()

# Combining height and width to determine overall size of trees
neighborhood_avg_size['overall_size'] = neighborhood_avg_size.mean(axis=1)

# Sorting neighborhoods by overall tree size in ascending order
sorted_size = neighborhood_avg_size['overall_size'].sort_values()

# Getting the neighborhoods with the smallest trees
smallest_10 = sorted_size.head(10)

# Plotting the neighborhoods with the least amount of trees
least_trees_10.plot(kind='bar', color='green', figsize=(10, 6))
plt.title('Neighborhoods with Least Amount of Trees')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Trees')
plt.show()

# Plotting the neighborhoods with the smallest trees (using height and width)
smallest_10.plot(kind='bar', color='orange', figsize=(10, 6))
plt.title('Neighborhoods with Smallest Trees (Avg Height & Width)')
plt.xlabel('Neighborhood')
plt.ylabel('Overall Size')
plt.show()
```

### Conclusion

The data that we got shows us the neighborhoods with the least amount of *recorded* trees and the neighborhood with the smallest trees. Given our overarching metric that the "best" neighborhood in Pittsburgh are neighborhoods that are best suited for younger people that thrive on convenience and cheap living, places that have a lower amount of trees or quality of trees would expect of have cheaper prices for housing. Regarding the amount of trees per neighborhood, the data shows that Hays, Glen Hazel, South Shore, Esplen, and Arlington Heights are the neighborhoods with the lowest *recorded* armount of trees. For the lowest quality (size) of trees, Allegheny Center, Perry North, Troy Hill, Highland Park, and Squirrel Hill South have the smallest trees on average. Given all of these factors, if trees themselves are the main factor, Hays would be considered the "best" neighborhood, but if the smallest trees are the main factor, Allegheney Center would be considered the "best" neighborhood.
