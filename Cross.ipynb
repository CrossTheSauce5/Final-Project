Link to dataset: https://www.drkattorneys.com/blog/neighborhood-crime/

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the dataset from the Google Sheets link
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT2tPJ6wWs68mu9-Kx1Gi-TiLUKrFYTumZBPdc3-zzirs1DWHXx5p_IxuRfehdN8ZjEDEYG8W6bgEWQ/pub?gid=0&single=true&output=csv"
df = pd.read_csv(url)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 10))  # Adjusted figsize for better visualization

# Finding neighborhoods with the highest amount of crimes
highest_violent_crime = df.nlargest(20, 'TOTAL VIOLENT CRIME')
highest_property_crime = df.nlargest(20, 'TOTAL PROPERTY CRIME')
highest_drug_crime = df.nlargest(20, 'TOTAL DRUG CRIME')

# Plotting neighborhoods with the highest amount of crimes
highest_violent_crime.plot(kind='barh', x='NEIGHBORHOOD', y='TOTAL VIOLENT CRIME', ax=axes[0], color='skyblue')
axes[0].set_title('Neighborhoods with Highest Violent Crime')
axes[0].set_xlabel('Total Violent Crime')

highest_property_crime.plot(kind='barh', x='NEIGHBORHOOD', y='TOTAL PROPERTY CRIME', ax=axes[1], color='salmon')
axes[1].set_title('Neighborhoods with Highest Property Crime')
axes[1].set_xlabel('Total Property Crime')

highest_drug_crime.plot(kind='barh', x='NEIGHBORHOOD', y='TOTAL DRUG CRIME', ax=axes[2], color='lightgreen')
axes[2].set_title('Neighborhoods with Highest Drug Crime')
axes[2].set_xlabel('Total Drug Crime')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the dataset from the Google Sheets link
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT2tPJ6wWs68mu9-Kx1Gi-TiLUKrFYTumZBPdc3-zzirs1DWHXx5p_IxuRfehdN8ZjEDEYG8W6bgEWQ/pub?gid=0&single=true&output=csv"
df = pd.read_csv(url)

# Calculate total crime count for each neighborhood
df['TOTAL CRIME'] = df['TOTAL VIOLENT CRIME'] + df['TOTAL PROPERTY CRIME'] + df['TOTAL DRUG CRIME']

# Sort neighborhoods by total crime count
df_sorted = df.sort_values(by='TOTAL CRIME', ascending=True)

# Plotting
plt.figure(figsize=(15, 20))
plt.barh(df_sorted['NEIGHBORHOOD'], df_sorted['TOTAL CRIME'], color='lightblue')
plt.xlabel('Total Crime Count')
plt.ylabel('Neighborhood')
plt.title('Total Crime Count by Neighborhood')
plt.gca().invert_yaxis()  # Invert y-axis to show the neighborhood with the highest crime count at the top
plt.grid(axis='x')  # Add gridlines only along the x-axis for better readability
plt.tight_layout()
plt.show()
```

# Conclusion

The data that we got for crime has two different ways to interpret the data. The dataset we evaluated graded these crimes like this: violent crimes are weighted at 0.6, property crimes at 0.3, and alcohol and drug crimes at 0.1. Because of the way they collected data, the neighborhood with the least amount of total violent crimes would have the highest grade. If we use this way of measuring, Regent Square would be the safest neighborhood. If we take into account all of the crimes (violent, property, drug) then the neighborhood with the least amount of crimes would still be Regent Square. If we only take into account the number of crimes, the best neighborhood to live in would be Regent Square.
